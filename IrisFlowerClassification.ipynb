```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

__LABEL CODING__

```python
iris['Species']= iris['Species'].astype('category')
```

```python
iris.dtypes
```

```python
iris['Species']= iris['Species'].cat.codes
```

```python
iris.head()
```

```python
iris['Species'].unique()
```

```python
iris.corr()
```

```python
X= iris.iloc[:,1:5]
```

```python
X.head()
```

```python
X.drop('SepalWidthCm', axis=1, inplace= True)
```

```python
X.head()
```

```python
X.to_numpy()
```

```python
y= iris['Species']
```

```python
y.head()
```

```python
y.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, train_size = 0.7, test_size = 0.3, random_state =  12)
```

__LOGISTIC REGRESSION MODEL__

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_Reg= LogisticRegression()
```

```python
Log_Reg.fit(X_train, y_train)
```

```python
y_pred_lg_train= Log_Reg.predict(X_train)
```

```python
y_pred_lg_test= Log_Reg.predict(X_test)
```

```python
score_lg_train = Log_Reg.score(X_train, y_train)
score_lg_test = Log_Reg.score(X_test, y_test)
print('Score_train:', score_lg_train)
print('Score_test:', score_lg_test)
```

```python
from sklearn import metrics
cm_test_lg = metrics.confusion_matrix(y_test, y_pred_lg_test)
print(cm_test_lg)
```

```python
cm_train_lg= metrics.confusion_matrix(y_train, y_pred_lg_train)
print(cm_train_lg)
```

```python
from sklearn.metrics import classification_report
print('Report_test:',classification_report(y_test, y_pred_lg_test))
```

```python
print('Report_train:',classification_report(y_train, y_pred_lg_train))
```
